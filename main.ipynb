```python
# Bibliotecas

# Manipulação de dados
import pandas as pd

# Faz a conversão de text para number
from sklearn.preprocessing import LabelEncoder

# Nearest Neighbors (KNN) > Vizinhos Proximos
from sklearn.neighbors import KNeighborsClassifier

# RandomForest > Arvore de Decisao
from sklearn.ensemble import RandomForestClassifier

# sklearn.metrics para verificar quão acurado são os modelos
from sklearn.metrics import accuracy_score

# Lightgbm
import lightgbm as lgb

# Se for usar o google drive (como vimos em aula)
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Importando base de dados do google drive (ajuste o caminho até seu arquivo)
df = pd.read_csv('drive/MyDrive/clientes.csv', sep=',')

# Descomente, caso queira visualizar a tabela de dados
# df.head(3)
```

#

# Pré-Processamento

É Nessa parte que fazemos a limpeza das base e deixamos em um formato que os modelos possam utilizar.

```python
# Vamos usar sklearn.preprocessing para converter as colunas necessárias de text para numbers
# Note: por ser um caso de laboratório a base já esta bem ajustada.

# Escrevi dessa forma, por conta da legibilidade, mas existem muitas opções para reescrever de forma otimizada

labelencoder_profissao = LabelEncoder()
df["profissao"] = labelencoder_profissao.fit_transform(df["profissao"])

labelencoder_mix_credito = LabelEncoder()
df["mix_credito"] = labelencoder_mix_credito.fit_transform(df["mix_credito"])

labelencoder_comportamento_pagamento = LabelEncoder()
df["comportamento_pagamento"] = labelencoder_comportamento_pagamento.fit_transform(df["comportamento_pagamento"])

# Visualizando os formatos dos dados
df.info()
```

#

```python
# Em y vamos ter os dados que iremos prever
y = df["score_credito"].copy()
```

```python
# Em x temos os dados que serão utilizados para treinar o modelo (id_cliente e score_credito podem ser omitidas, uma vez que não usaremos)
columns_model = [
    'mes', 'idade', 'profissao', 'salario_anual', 'num_contas',
    'num_cartoes', 'juros_emprestimo', 'num_emprestimos',
    'dias_atraso', 'num_pagamentos_atrasados', 'num_verificacoes_credito',
    'mix_credito', 'divida_total', 'taxa_uso_credito',
    'idade_historico_credito', 'investimento_mensal',
    'comportamento_pagamento', 'saldo_final_mes',
    'emprestimo_carro', 'emprestimo_casa', 'emprestimo_pessoal',
    'emprestimo_credito', 'emprestimo_estudantil'
]

x = df[columns_model].copy()
```

```python
# Biblioteca utilizada para separar a base de treino e base de teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

# Vamos testar nossa base com alguns modelos de ML

[Link para biblioteca: Scikit Learn](https://scikit-learn.org/stable/supervised_learning.html)

# Random Forest

```python
# Criação do modelo
modelo_arvoredecisao = RandomForestClassifier()

# Treinamento
modelo_arvoredecisao.fit(x_treino, y_treino)
```

#

# KNN (K-Nearest Neighbors)

```python
# Criação do modelo
modelo_knn = KNeighborsClassifier()

# Treinamento
modelo_knn.fit(x_treino, y_treino)
```

<br>

# LIGHTGMB

```python
# Criação do modelo (existe uma grande gama de parâmetros que podemos alterar no modelo)
modelo_lightgmb = lgb.LGBMClassifier(n_estimators=200, learning_rate=0.03)

# Treinamento
modelo_lightgmb.fit(x_treino, y_treino)
```

<br>

### Vamos utilizar os dados que deixamos reservado para comparar o desempenho.

```python
# verificando as previsões com base nos dados de teste
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)
previsao_lightgmb = modelo_lightgmb.predict(x_teste)
```

```python

print(f'Random Forest ==> {accuracy_score(y_teste, previsao_arvoredecisao)}')
print(f'Nearest Neighbors (KNN) ==> {accuracy_score(y_teste, previsao_knn)}')
print(f'Lightgmb (Lightgmb) ==> {accuracy_score(y_teste, previsao_lightgmb)}')
```

<br>

Verificamos que pra esse problema o modelo de arvore de decisão teve o melhor desempenho. Poderíamos fazer várias modificações ou testar outros para tentar ter mais performance.

Agora podemos usar o melhor modelo para classificar novos clientes

```python
# novos_clientes (ajuste o caminho até seu arquivo)
df_novos_clientes = pd.read_csv("drive/MyDrive/clientesNovos.csv", sep=',')

# Pré processamento (igual fizemos com a base de treino)
df_novos_clientes["profissao"] = labelencoder_profissao.transform(df_novos_clientes["profissao"])
df_novos_clientes["mix_credito"] = labelencoder_mix_credito.transform(df_novos_clientes["mix_credito"])
df_novos_clientes["comportamento_pagamento"] = labelencoder_comportamento_pagamento.transform(df_novos_clientes["comportamento_pagamento"])

# Descomente, caso queira visualizar a tabela de novos clientes
df_novos_clientes
```

```python
# Calculando os score de crédito com base no modelo KNN
score_previsao = modelo_arvoredecisao.predict(df_novos_clientes)
print(score_previsao)
```

```python
# Adicionando a coluna com os scores para os novos clientes
df_novos_clientes['score_credito'] = score_previsao

df_novos_clientes
```
